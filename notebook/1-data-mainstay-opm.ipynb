# Mainstay OPM

```python
import pandas as pd
import numpy as np
```

```python
# Read the dataset
mainstay_df = pd.read_csv('../data/mainstay_tracks17to23.csv')
tracks_df = pd.read_csv('../data/ph_spotify_daily_charts_tracks.csv')
artists_df = pd.read_csv('../data/ph_spotify_daily_charts_artists.csv')
```

```python
# Get the relevant columns in the spotify daily charts tracks dataframe
tracks_df = tracks_df[['track_id', 'artist_id', 'duration', 'danceability', 'energy', 'key', 'loudness',
                       'mode', 'speechiness', 'acousticness', 'instrumentalness', 'liveness', 'valence', 'tempo']]
```

```python
# Get the relevant columns in the spotify daily charts artists dataframe
artists_df = artists_df[['artist_name', 'genres']]
```

```python
# Merge the mainstay and tracks dataframes on 'track_id'
merged_df = pd.merge(mainstay_df, tracks_df, on='track_id', how='left')
merged_df
```

```python
# Merge the merged dataframe and artists dataframe on 'artist_name'
final_df = pd.merge(merged_df, artists_df, on='artist_name', how='left')
final_df
```

```python
# Append the new track to final_df
new_track = {'track_id': '47OFMuRu5lzJHRbtEXPBbs', 'track_name': 'Uhaw (Tayong Lahat)', 'artist_name': 'Dilaw',
             'streams': 90213, 'artist_id': '6Dp4LInLyMVA2qhRqQ6AGL', 'total_followers': 38283, 'popularity': 79,
             'duration': 241185, 'danceability': 0.635, 'energy': 0.724, 'key': 5.0, 'loudness': -6.662,
             'mode': 1.0, 'speechiness': 0.0716, 'acousticness': 0.28, 'instrumentalness': 0.000004,
             'liveness': 0.105, 'valence': 0.316, 'tempo': 135.927, 'genres': 'opm'}
final_df = final_df.append(new_track, ignore_index=True)
final_df
```

```python
final_df.drop(['total_followers', 'popularity'], axis=1, inplace=True)
final_df
```
