# Uhaw Virality
> Viral is defined as those tracks that achieve a minimum of 50,000 daily streams and exhibit a significant decrease in popularity, as demonstrated by a minimum daily rank drop of 35 positions and a daily stream decrease of at least 10,000.

```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
# read and process the charts dataset
df = pd.read_csv('data/ph_spotify_daily_charts.csv')

#transform date column into a datetime column
df['date'] = pd.to_datetime(df['date'])
df = df.set_index('date')
df.head()
```

```python
df['track_name'].nunique()
```

```python
df.columns
```

```python
# Code here
dilaw = df[df['artist']=='Dilaw']
dilaw
```

```python
dilaw = dilaw.copy()
dilaw["streams_increase"] = dilaw.groupby("track_name")["streams"].diff()
dilaw["position_change"] =  -dilaw.groupby("track_name")["position"].diff() #- sign because the higher the number the lower its ranking
dilaw
```

```python
X2 = 50000  # 100k streams/day
P2 = -35  # 35 positions/day
X3 = -10000  # 20k streams/day
```

```python
dilaw[dilaw["streams"] > X2]
```

```python
dilaw[(dilaw["position_change"] < P2)]
```

```python
dilaw[dilaw["streams_increase"] < X3]
```

```python
is_viral = ((dilaw["streams"] > X2) & (dilaw["position_change"] < P2) & (dilaw["streams_increase"] < X3)) 
is_viral
```

```python
dilaw["viral"] = is_viral
dilaw
```

```python
# dilaw.to_csv("data/dilaw.csv", index=False)
```

```python
viral_tracks = dilaw[dilaw["viral"]]
viral_tracks
```

```python
chosen_list = ['Uhaw', 'Uhaw (Tayong Lahat)'] 
```

```python
fig, ax = plt.subplots(figsize=(13,4))

for track in chosen_list:
    data = df[df['track_name'] == track]['streams']
    data.plot(ax=ax, label = track)
    
    plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.ylabel('streams')
    plt.title('Dilaw Streams')
    
```

The trend for the streams is generally upwards for both songs. 

```python
fig, ax = plt.subplots(figsize=(13,4))

for track in chosen_list:
    data = df[df['track_name'] == track]['position']
    data.plot(ax=ax, label = track)
    
    plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.ylabel('position')
    plt.title('Dilaw Song Positions')
    
```

The trend for the position is generally downwards for both songs.

However, the song maintained a steady position in the top 100. Additionally, the song did not experience a significant decrease in daily streams during the time period.

Therefore, it can be concluded that "Uhaw (Tayong Lahat)" and Uhaw by Dilaw did not meet the criteria for virality.
